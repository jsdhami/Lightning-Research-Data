```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as math
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/jsdhami/Lightning-Research-Data/main/Data/GD/TIA%20Data.csv')
df=pd.DataFrame({'date':df['date'],'min_temp':df['tmin'],'max_temp':df['tmax'],'ave_temp':df['tavg'],'air_pressure':df['pres'],'precipitation':df['prcp']})
df.head()


```

```python
w=df['ave_temp']
x=df['date']
y=df['air_pressure']
z=df['precipitation']
```

```python
#fig plot
fig,ax=plt.subplots()
fig.set_size_inches(10,5)
ax.plot(x,y,label='Air Pressure',color='green')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Air Pressure (hpa)',fontsize=15)
ax.set_title('Air Pressure in Kathmandu Airport',fontsize=20)
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=8)
plt.style.use('default')
ax.legend()
plt.show()
```

```python
#Line plot
fig,ax=plt.subplots()
fig.set_size_inches(10,5)
ax.plot(x,z,label='Precipitation',color='blue')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Precipitation (mm)',fontsize=15)
ax.set_title('Precipitation in Kathmandu Airport',fontsize=20)
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=8)
plt.style.use('default')
ax.legend()
plt.show()
```

```python
#precipitation vs air pressure
fig,ax=plt.subplots()
fig.set_size_inches(15,7)
#Plot
ax.plot(x,y,label='Air pressure',color='green')

#create twinx
ax2=ax.twinx()
ax2.plot(x,z,label='Precipitation',color='blue')
#set labelling
fig.suptitle('Air pressure and precipitation in TIA KTM',fontsize=20,fontweight='bold')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Air Pressure (hpa)',fontsize=15,fontweight="bold")
ax2.set_ylabel('Precipitation (mm)',fontsize=15,fontweight="bold")
#set ticks and ticks label
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=15)
plt.style.use('default')
#Legend
legend_properties = {'weight':'bold','size':'15'}
ax.legend(loc='upper left',prop=legend_properties)
ax2.legend(loc='upper right',prop=legend_properties)
plt.show()
```

```python
#air pressure vs temperature
fig,ax=plt.subplots()
fig.set_size_inches(15,7)
#Plot
ax.plot(x,w,label='Temperature',color='red')
#twinx
ax2=ax.twinx()
ax2.plot(x,y,label='Air pressure',color='green')
#Labelling
ax.set_title('Air pressure and temperature in TIA KTM',fontsize=20,fontweight='bold')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Temperature (C)',fontsize=15,fontweight="bold")
ax2.set_ylabel('Air Pressure (hpa)',fontsize=15,fontweight="bold")
#set ticks and ticks label
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=15)
#style
plt.style.use('default')
#Legend
legend_properties = {'weight':'bold','size':'15'}
ax.legend(loc='upper left',prop=legend_properties)
ax2.legend(loc='upper right',prop=legend_properties)
plt.show()
```
