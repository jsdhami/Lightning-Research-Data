```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math as math
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jsdhami/Lightning-Research-Data/main/Data/GD/TIA%20Data.csv')
df = pd.DataFrame({'Date': df['date'], 'Min temp': df['tmin'], 'Max temp': df['tmax'], 'Avg temp': df['tavg'], 'Air pressure': df['pres'], 'precipitation': df['prcp']})
df.head()
```

```python
# creating variable
w = df['Avg temp']
x = df['Date']
y = df['Air pressure']
z = df['precipitation']
```

# **Ploting**

---



```python
# Line plot of Air pressure vs Date
fig,ax=plt.subplots()
fig.set_size_inches(10,5)
ax.plot(x,y,label='Air pressure',color='yellow')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Air presure(hectapascal)',fontsize=15)
ax.set_title('Air pressure in Kathmandu Airport',fontsize=20)
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=8)
plt.style.use('dark_background')
ax.legend()
plt.show()
ax.plot()
```

```python
#Line plot of precipitation vs date in Kathmandu airport
fig,ax=plt.subplots()
fig.set_size_inches(10,5)
ax.plot(x,z,label='Precipitation',color='red')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Precipitation (mm)',fontsize=15)
ax.set_title('Precipitation in Kathmandu Airport',fontsize=20)
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=8)
plt.style.use('dark_background')
ax.legend()
plt.show()
```

```python
#precipitation and air pressure in one plot vs date
fig,ax=plt.subplots()
fig.set_size_inches(15,7)
#Plot
ax.plot(x,y,label='Air pressure',color='yellow')

#create twinx
ax2=ax.twinx()
ax2.plot(x,z,label='Precipitation',color='Red')
#set labelling
fig.suptitle('Air pressure and precipitation in TIA KTM',fontsize=20,fontweight='bold')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Air Pressure (hpa)',fontsize=15,fontweight="bold")
ax2.set_ylabel('Precipitation (mm)',fontsize=15,fontweight="bold")
#set ticks and ticks label
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=15)
plt.style.use('dark_background')
#Legend
legend_properties = {'weight':'bold','size':'15'}
ax.legend(loc='upper left',prop=legend_properties)
ax2.legend(loc='upper right',prop=legend_properties)
plt.show()
```

```python
#air pressure vs temperature in TIA
fig,ax=plt.subplots()
fig.set_size_inches(15,7)
#Plot
ax.plot(x,w,label='Temperature',color='green')
#twinx
ax2=ax.twinx()
ax2.plot(x,y,label='Air pressure',color='yellow')
#Labelling
ax.set_title('Air pressure and temperature in TIA KTM',fontsize=20,fontweight='bold')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Temperature (C)',fontsize=15,fontweight="bold")
ax2.set_ylabel('Air Pressure (hpa)',fontsize=15,fontweight="bold")
#set ticks and ticks label
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=15)
#style
plt.style.use('default')
#Legend
legend_properties = {'weight':'bold','size':'15'}
ax.legend(loc='upper left',prop=legend_properties)
ax2.legend(loc='upper right',prop=legend_properties)
plt.show()
```
