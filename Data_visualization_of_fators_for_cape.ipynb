```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/jsdhami/Lightning-Research-Data/main/Daily_Data_CAPE_LST.csv')
```

```python
#Adding new coloum date
df['Date'] = df['YEAR'].astype(str) + '-'+ df['MO'].astype(str) + '-' + df['DY'].astype(str)
df=df[df['T2M']!=-999.0]
df.head()
```

```python
#Creating variables
x = df['Date']
y = df['T2M']
z = df['T2MDEW']
u = df['QV2M']
v = df['RH2M']
w = df['WS10M']
m = df['WD10M']
n = df['PSC']
o = df['WSC']
#T2M = temperature at 2 meter height
#T2MDEW = Temperature of dew point at 2m
#QV2M = Specific humidity at 2m
#RH2M = Relative humidity at 2m
#PS = Surface pressure
#WS10M = wind speed at 10 m
#WD10M = Wind direction at 10 m
#PSC = Corrected surface pressure (Corrected for site elevation)
#WSC = corrected wind speed (Corrected for site elevation)
```

```python
x.head()
o.min()
```

# **Ploting**

---

Line ploting

```python
# Line plot of T2M vs Date
fig,ax=plt.subplots()
fig.set_size_inches(10,5)
ax.plot(x,y,label='Temperature',color='Red')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Temperature',fontsize=15)
ax.set_title('Temperature of kathmandu at 2m height',fontsize=20)
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=8)
plt.style.use('dark_background')
ax.legend()
plt.show()
ax.plot()
```

```python
# Line plot of WSC vs Date
fig,ax=plt.subplots()
fig.set_size_inches(15,7)
ax.plot(x,o,label='Corrected wind speed',color='Blue')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Corrected wind speed',fontsize=15)
ax.set_title('Corrected wind speed',fontsize=20)
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=8)
plt.style.use('dark_background')
ax.legend()
plt.show()
ax.plot()
```

```python
# Line plot of PSC vs Date
fig,ax=plt.subplots()
fig.set_size_inches(15,7)
ax.plot(x,n,label='Corrected surface pressure',color='Green')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Corrected Surface presssure',fontsize=15)
ax.set_title('Corrected Surface presure',fontsize=20)
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=8)
plt.style.use('dark_background')
ax.legend()
plt.show()
ax.plot()
```

```python
#Relative humidity vs specific huidty
fig,ax=plt.subplots()
fig.set_size_inches(15,7)
#Plot
ax.plot(x,u,label='Specific humidity',color='green')
#twinx
ax2=ax.twinx()
ax2.plot(x,v,label='relative humidity',color='red')
#Labelling
ax.set_title('Specific humidity vs relative humidity',fontsize=20,fontweight='bold')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Specific humidity at 2 m',fontsize=15,fontweight="bold")
ax2.set_ylabel('Relative humidity at 2 m',fontsize=15,fontweight="bold")
#set ticks and ticks label
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=15)
#style
plt.style.use('dark_background')
#legend
legend_properties = {'weight':'bold','size':'15'}
ax.legend(loc='upper left',prop=legend_properties)
ax2.legend(loc='upper right',prop=legend_properties)
#show
plt.show()
```

```python
#Windspeed vs wind direction at 10 m
fig,ax=plt.subplots()
fig.set_size_inches(15,7)
#Plot
ax.plot(x,w,label='Wind speed',color='yellow')
#twinx
ax2=ax.twinx()
ax2.plot(x,m,label='Wind direction',color='pink')
#Labelling
ax.set_title('Wind speed vs Wind direction ',fontsize=20,fontweight='bold')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Wind speed at 10 m',fontsize=15,fontweight="bold")
ax2.set_ylabel('Wind direction at 10 m',fontsize=15,fontweight="bold")
#set ticks and ticks label
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=15)
#style
plt.style.use('dark_background')
#legend
legend_properties = {'weight':'bold','size':'15'}
ax.legend(loc='upper left',prop=legend_properties)
ax2.legend(loc='upper right',prop=legend_properties)
#show
plt.show()
```

```python
# Scatter plot of T2MDEW vs Date
fig,ax=plt.subplots()
fig.set_size_inches(10,5)
ax.scatter(x,y,label='Temperature of DEW point',color='yellow')
ax.set_xlabel('Date',fontsize=15)
ax.set_ylabel('Temperature of Dew point',fontsize=15)
ax.set_title('Temperature of Dew point at 2m height',fontsize=20)
ax.set_xticks(x[::20])
ax.set_xticklabels(x[::20],rotation=90,fontsize=8)
plt.style.use('dark_background')
ax.legend()
plt.show()
ax.plot()
```
