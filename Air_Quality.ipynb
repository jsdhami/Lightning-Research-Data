````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
````

````python
# read csv datta
AirQuality = pd.read_csv('https://raw.githubusercontent.com/jsdhami/Lightning-Research-Data/main/Data/JS/KTM-AirQualityData-7-20to7-30.csv')
AirQuality.head()
````

````python
New_DF = pd.DataFrame(
    {
        "locationId":AirQuality['locationId'],
        "pm25":AirQuality[AirQuality['parameter']=='pm25']['value'],
        "pm10":AirQuality[AirQuality['parameter']=='pm10']['value'],
        "pm1":AirQuality[AirQuality['parameter']=='pm1']['value'],
        "temperature":AirQuality[AirQuality['parameter']=='temperature']['value'],
        "relativehumidity":AirQuality[AirQuality['parameter']=='relativehumidity']['value'],
        "datetime":AirQuality['date']
    }
)

New_DF.head()
````

````python
# for pm25
pm25_df = pd.DataFrame(
    {
        "locationId":AirQuality[AirQuality['parameter']=='pm25']['locationId'],
        "location":AirQuality[AirQuality['parameter']=='pm25']['location'],
        "pm25":AirQuality[AirQuality['parameter']=='pm25']['value'],
        "unit":AirQuality[AirQuality['parameter']=='pm25']['unit'],
        "datetime":AirQuality[AirQuality['parameter']=='pm25']['date'],
        "coordinates": AirQuality[AirQuality['parameter']=='pm25']['coordinates']
    }
)
pm25_df.head(555)
````
